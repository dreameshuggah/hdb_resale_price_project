```python
import os
from pathlib import Path

print(os.getcwd())
raw_data_path = Path("/Volumes/SANDISKUSBC/MLOps/hdb_price_prediction_mlops/data/raw")
os.listdir(raw_data_path)
```

```python
import json
import pandas as pd
from urllib.request import Request, urlopen

hdr = {'User-Agent': 'Mozilla/5.0', "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,/;q=0.8"}
url = "https://tablebuilder.singstat.gov.sg/api/table/tabledata/M212161?isTestApi=true"

# Fetch the data
request = Request(url, headers=hdr)
data_bytes = urlopen(request).read()

# Parse the bytes data into a JSON dictionary
json_data = json.loads(data_bytes.decode('utf-8'))

# Safely extract the 'row' list
records = json_data.get('Data', {}).get('row', [])

if records:
    # Fetch the 'columns' data from the very first record
    columns_data = records[0].get('columns', [])
    
    # Let's print the first 5 records to verify
    print("First 5 records from 'columns':")
    print(json.dumps(columns_data[:5], indent=4))
    
    # Optional: Convert it cleanly into a Pandas DataFrame!
    # Rename 'key' to 'Quarter' and 'value' to 'Resale_Price_Index'
    df = pd.DataFrame(columns_data).rename(columns={'key': 'quarter', 'value': 'resale_price_index'})
    
    # Convert the value column to float (it comes in as a string initially)
    df['resale_price_index'] = pd.to_numeric(df['resale_price_index'], errors='coerce')
    
    print("\nDataFrame Output:")
    display(df.head())
else:
    print("No records found in the data.")
```
